```python
df.loc[ df['age'] <= 10, 'age_clean'] = 0
df.loc[(df['age'] > 10) & (df['age'] <= 20), 'age_clean'] = 1
df.loc[(df['age'] > 20) & (df['age'] <= 30), 'age_clean'] = 2
df.loc[(df['age'] > 30) & (df['age'] <= 40), 'age_clean'] = 3
df.loc[(df['age'] > 40) & (df['age'] <= 50), 'age_clean'] = 4
df.loc[(df['age'] > 50) & (df['age'] <= 60), 'age_clean'] = 5
df.loc[(df['age'] > 60) & (df['age'] <= 70), 'age_clean'] = 6
df.loc[(df['age'] > 70) & (df['age'] <= 80), 'age_clean'] = 7
df.loc[(df['age'] > 80) & (df['age'] <= 90), 'age_clean'] = 8
df.loc[(df['age'] > 90) & (df['age'] <= 100), 'age_clean'] =9

```
